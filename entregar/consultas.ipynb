# Consulta 5

```python
%%sql
CREATE VIEW venta_por_producto AS select op.rutEmprendedor, op.idProducto, count(*) 
from s.ofreceProducto op, s.compraProducto cp 
where (op.idProducto=cp.idProducto) group by op.rutEmprendedor, op.idProducto;
```

```python
%sql select * from venta_por_producto order by rutEmprendedor;
```

```python
%%sql 
select vpp.rutEmprendedor, max(vpp.count) as maximo_por_vendedor
from venta_por_producto vpp where vpp.rutEmprendedor=vpp.rutEmprendedor group by vpp.rutEmprendedor;
```

# Consulta 6

```python
%%sql
CREATE VIEW venta_por_servicio AS select os.rutEmprendedor, os.idServicio, count(*) 
from s.ofreceServicio os, s.compraServicio cs 
where (os.idServicio=cs.idServicio) group by os.rutEmprendedor, os.idServicio;
```

```python
%sql select * from venta_por_servicio order by rutEmprendedor;
```

```python
%%sql 
select vps.rutEmprendedor, max(vps.count) as maximo_por_vendedor
from venta_por_servicio vps where vps.rutEmprendedor=vps.rutEmprendedor group by vps.rutEmprendedor;
```

# Vista 1

```python
%%sql
create or replace view descuento(cliente,mes, totalPagado, nroCompras) as
select c.nombrecliente, p.montoNeto, count(*)
from Cliente c, Maestro m, compraProducto p,compraServicio s
where (c.rutCliente=s.rutCliente and s.idMaestro=m.idMaestro and m.fecha=MONTH(GET DATE()) ) or (c.rutCliente=p.rutCliente and p.idMaestro=m.idMaestro and m.fecha=MONTH(GET DATE()))
group by m.idMaestro;
select * from descuento;
select cliente, mes, nroCompras, totalPagado
where totalPagado= (select max(c2.totalpagado)from descuento c2);
```

## Vendedores

```python
%%sql 
CREATE OR REPLACE FUNCTION s.ingresar_vendedor(VARCHAR,
	VARCHAR,
	VARCHAR,
	integer,
	VARCHAR) 
RETURNS void AS $$
INSERT INTO s.Emprendedor(rutEmprendedor,
	nombreEmprendedor,
	paginaWeb,
	telefonoEmprendedor,
	correoEmprendedor) values($1,
	$2,
	$3,
	$4,
	$5) ;
$$ LANGUAGE SQL;
```

```python
%%sql
CREATE OR REPLACE FUNCTION s.modificar_vendedor(VARCHAR,
	VARCHAR,
	VARCHAR,
	integer,
	VARCHAR) 
RETURNS void AS $$
UPDATE s.Emprendedor
SET nombreEmprendedor = $2,
	paginaWeb = $3,
	telefonoEmprendedor = $4,
	correoEmprendedor = $5
where rutEmprendedor = $1;
$$ LANGUAGE SQL;
```

```python
%%sql
CREATE OR REPLACE FUNCTION s.eliminar_vendedor(VARCHAR)
RETURNS void AS $$
DELETE from s.Emprendedor where rutEmprendedor=$1;
$$ LANGUAGE SQL;
```

```python
CREATE OR REPLACE FUNCTION restringir_edad()
RETURNS TRIGGER AS $$
DECLARE num INTEGER;
BEGIN
	num:=(SELECT COUNT(*) FROM pf.cartera_vendedor WHERE nombre_vendedor=NEW.nombre_vendedor);
	IF num>=18 THEN
		RAISE EXCEPTION 'Usted es menor de edad';
END; $$
LANGUAGE plpgsql;
```
